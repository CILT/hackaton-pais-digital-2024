# Introducción a OpenCV

OpenCV (Open Source Computer Vision Library) es una biblioteca de código abierto que contiene más de 2500 algoritmos optimizados, que permiten realizar tareas de visión por computadora en tiempo real. OpenCV es ampliamente utilizado para aplicaciones como detección de rostros, reconocimiento de objetos, clasificación de imágenes, y más.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## 1. Cargar y visualizar una imagen
Vamos a comenzar cargando una imagen y mostrándola utilizando OpenCV y matplotlib:

```python
# Cargar la imagen
image = cv2.imread('Rocky.jpg')

# Convertir de BGR (formato predeterminado de OpenCV) a RGB
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Mostrar la imagen usando matplotlib
plt.imshow(image_rgb)
plt.title('Imagen Original')
plt.axis('off')
plt.show()
```

- Observación: OpenCV carga las imágenes en formato BGR, mientras que otras bibliotecas de visualización, como matplotlib, utilizan el formato RGB. Por eso, convertimos la imagen antes de mostrarla.

## 2. Operaciones Básicas en Imágenes
### 2.1. Escalado
Escalar una imagen es una de las operaciones más básicas y se puede hacer fácilmente con `cv2.resize`:

```python
# Redimensionar la imagen al 50% de su tamaño original
image_scaled = cv2.resize(image, None, fx=0.5, fy=0.5)

# Mostrar la imagen escalada
plt.imshow(cv2.cvtColor(image_scaled, cv2.COLOR_BGR2RGB))
plt.title('Imagen Escalada al 50%')
plt.axis('off')
plt.show()
```

Como la imagen es de gran tamaño, no se nota a simple vista el escalado, pero se pueden revisar directamente las dimensiones de las imágenes.

```python
# Obtener dimensiones de la imagen original
original_height, original_width = image.shape[:2]
print(f'Tamaño de la imagen original: {original_width}x{original_height}')

# Obtener dimensiones de la imagen redimensionada
scaled_height, scaled_width = image_scaled.shape[:2]
print(f'Tamaño de la imagen escalada: {scaled_width}x{scaled_height}')
```

Un ejemplo para observar el redimensionamiento, es escalar la imagen al 1% de su tamaño original.

```python
# Redimensionar la imagen al 1% de su tamaño original
image_scaled = cv2.resize(image, None, fx=0.01, fy=0.01)

# Mostrar la imagen escalada
plt.imshow(cv2.cvtColor(image_scaled, cv2.COLOR_BGR2RGB))
plt.title('Imagen Escalada al 1%')
plt.axis('off')
plt.show()
```

### 2.2 Conversión a Escala de Grises

Para algunas aplicaciones resulta útil trabajar únicamente con un canal en vez de 3 canales por imagen. 

```python
# Convertir la imagen a escala de grises
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Mostrar la imagen en escala de grises
plt.imshow(image_gray, cmap='gray')
plt.title('Imagen en Escala de Grises')
plt.axis('off')
plt.show()
```

### 2.3. Detección de Bordes
Podemos utilizar el algoritmo de Canny para la detección de bordes.

```python
# Aplicar el detector de bordes Canny
edges = cv2.Canny(image_gray, threshold1=100, threshold2=200)

# Mostrar los bordes detectados
plt.imshow(edges, cmap='gray')
plt.title('Detección de Bordes con Canny')
plt.axis('off')
plt.show()
```

## 3. Dibujar Formas en Imágenes

### 3.1. Dibujar un Rectángulo

La función cv2.rectangle permite dibujar un rectángulo en una imagen en base a 5 parámetros:

1. La imagen sobre la que se va a dibujar el rectángulo.
2. Las coordenadas (x, y) de la esquina superior izquierda del rectángulo.
3. Las coordenadas (x, y) de la esquina inferior derecha del rectángulo.
4. El color de la línea del rectángulo en formato BGR (Blue, Green, Red).
5. El grosor de la línea que forma los bordes del rectángulo. Si el grosor toma el valor de -1, se asume que el rectángulo debe ser rellenado completamente con el color especificado.


```python
# Dibujar un rectángulo en la imagen original
image_rectangle = image.copy()
cv2.rectangle(image_rectangle, (50, 100), (2000, 1200), (255, 255, 0), 20)

# Convertir la imagen de BGR a RGB (OpenCV usa BGR por defecto)
image_rgb = cv2.cvtColor(image_rectangle, cv2.COLOR_BGR2RGB)

# Mostrar la imagen con el rectángulo
plt.imshow(image_rgb)
plt.title('Dibujo de Rectángulo')
plt.axis('off')
plt.show()
```

### 3.2. Escribir Texto

La función `cv2.putText` permite escribir texto en una imagen en base a 8 parámetros:

1. La imagen sobre la que se va a dibujar el texto.
2. El texto que se desea dibujar en la imagen.
3. Las coordenadas (x, y) de la esquina inferior izquierda del texto.
4. El tipo de fuente que se utilizará para el texto (por ejemplo, cv2.FONT_HERSHEY_SIMPLEX).
5. El tamaño de la fuente (un factor de escala que afecta la altura y anchura del texto).
6. El color del texto en formato BGR (Blue, Green, Red).
7. El grosor de las líneas que forman las letras del texto.
8. El tipo de línea utilizado para dibujar el texto (por ejemplo, cv2.LINE_AA para aplicar antialiasing).


```python
# Escribir texto en la imagen
image_text = image.copy()
cv2.putText(image_text, 'Hackaton 2024', (300, 300), cv2.FONT_HERSHEY_SIMPLEX, 10, (255, 255, 0), 20, cv2.LINE_AA)

# Convertir la imagen de BGR a RGB (OpenCV usa BGR por defecto)
image_rgb = cv2.cvtColor(image_text, cv2.COLOR_BGR2RGB)

# Mostrar la imagen con el texto
plt.imshow(image_rgb)
plt.title('Escribir texto en una imagen')
plt.axis('off')
plt.show()
```
